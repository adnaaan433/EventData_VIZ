```python
frwd = df[(df['position']=='FW')]
fwd_list = frwd.name.unique()
fwd_list
```

```python
def forward_ball_retention(pname):
    ball_retention = None
    m_played = df[(df['name'] == pname) & (df['position'] == 'FW')]
    team_name = m_played['teamName'].max()
    # Check if player has played 4 or more matches
    if len(m_played['match_id'].unique()) >= 4:
        next_act = df[(df['name'] == pname) & (df['type'].shift(1) == 'Pass') & (df['outcomeType'].shift(1) == 'Successful')]
        
        # Check if player has received 50 or more passes
        if len(next_act) >= 50:
            ball_retain = next_act[(next_act['outcomeType'] == 'Successful') & (~next_act['type'].str.contains('Dispossessed|Foul'))]
            ball_retention = round((len(ball_retain) / len(next_act)) * 100, 2)
        else:
            ball_retention = 0

    return ball_retention, team_name

# List to hold the retention data
retention_data = []
for player in fwd_list:
    retention, team_name = forward_ball_retention(player)
    # Only append data if ball_retention is not None
    if retention is not None:
        retention_data.append({'name': player, 'ball_retention': retention, 'team_name': team_name})

# Convert the list into a DataFrame
ball_retention_df = pd.DataFrame(retention_data)

# Sort and reset index
ball_retention_df = ball_retention_df.sort_values(by='ball_retention', ascending=False).reset_index(drop=True)
ball_retention_df
```
