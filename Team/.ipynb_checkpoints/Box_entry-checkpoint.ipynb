```python
def plot_blank(ax):
    pitch = VerticalPitch(pitch_type='uefa',pitch_color='#f5f5f5',line_color='#f5f5f5',corner_arcs=True,line_zorder=3)
    pitch.draw(ax=ax)
```

Pass

```python
def pass_entry_for(ax,team_name):
    df_team = df[(df['teamName']==team_name) | (df['oppositionTeamName']==team_name)]
    df_team['next_qualifiers'] = df_team['qualifiers'].shift(-1)
    df_team['Length'] = np.sqrt((df['x']-df['endX'])**2 + (df['y']-df['endY'])**2)
    df_team = df_team[df_team['teamName']==team_name]
    
    box_entries = df_team[((df_team['type']=='Pass')|(df_team['type']=='Carry')) & (df_team['outcomeType']=='Successful') & (df_team['endX']>=88.5) &
                 ~((df_team['x']>=88.5) & (df_team['y']>=13.6) & (df_team['y']<=54.6)) & ((df_team['Length']>=5) | (df_team['Length']>=5)) &
                 (df_team['endY']>=13.6) & (df_team['endY']<=54.4) & (~df_team['qualifiers'].str.contains('CornerTaken|Freekick|ThrowIn'))]
    df_rigt = box_entries[(box_entries['type']=='Pass') & (box_entries['y']<=68/3)]
    df_cent = box_entries[(box_entries['type']=='Pass') & (box_entries['y']>68/3) & (box_entries['y']<=136/3)]
    df_left = box_entries[(box_entries['type']=='Pass') & (box_entries['y']>136/3)]
    pitch = VerticalPitch(pitch_type='uefa',pitch_color='#f5f5f5',line_color='k',corner_arcs=True,line_zorder=3)
    pitch.draw(ax=ax)
    ax.set_ylim(52.25, 106)

    t=s=g=0
    for index, row in box_entries.iterrows():
        if (row['type']=='Pass'):
            arrow = patches.FancyArrowPatch((row['y'], row['x']), (row['endY'], row['endX']), arrowstyle='->',
                                        color='k', zorder=4, mutation_scale=10, alpha=0.15, linewidth=1)
            t+=1
            ax.add_patch(arrow)
            ax.scatter(row['y'], row['x'], color='k', edgecolor='w', alpha=0.15, zorder=4)
        if (row['type']=='Pass' and ('KeyPass' in row['qualifiers'] or 'KeyPass' in row['next_qualifiers'])):
            arrow = patches.FancyArrowPatch((row['y'], row['x']), (row['endY'], row['endX']), arrowstyle='->',
                                        color=col2, zorder=5, mutation_scale=10, alpha=0.6, linewidth=1.25)
            s+=1
            ax.add_patch(arrow)
            ax.scatter(row['y'], row['x'], color=col2, edgecolor='w', zorder=5)
        if (row['type']=='Pass' and ('IntentionalGoalAssist' in row['qualifiers'] or 'IntentionalGoalAssist' in row['next_qualifiers'])):
            arrow = patches.FancyArrowPatch((row['y'], row['x']), (row['endY'], row['endX']), arrowstyle='->',
                                        color='green', zorder=6, mutation_scale=15, alpha=1, linewidth=2)
            g+=1
            ax.add_patch(arrow)
            ax.scatter(row['y'], row['x'], color='green', edgecolor='w', zorder=6)

    left_counts = df_left['shortName'].value_counts()
    max_player_left = left_counts.idxmax()
    max_count_left = left_counts.max()
    ax.text(68, 47, f"Most from left:\n{max_player_left}:{max_count_left}", color='k', fontsize=18, va='center', ha='left')
    cent_counts = df_cent['shortName'].value_counts()
    max_player_cent = cent_counts.idxmax()
    max_count_cent = cent_counts.max()
    ax.text(34, 47, f"Most from center:\n{max_player_cent}:{max_count_cent}", color='k', fontsize=18, va='center', ha='center')
    rigt_counts = df_rigt['shortName'].value_counts()
    max_player_rigt = rigt_counts.idxmax()
    max_count_rigt = rigt_counts.max()
    ax.text(0, 47, f"Most from right:\n{max_player_rigt}:{max_count_rigt}", color='k', fontsize=18, va='center', ha='right')

    ax.text(60, 117, f'{team_name}', fontsize=35, fontweight='bold', ha='left', va='center')
    ax.text(68, 55, f'Box entries\nfrom left: {len(df_left)}', fontsize=18, ha='left', va='center', zorder=7,
            bbox=dict(facecolor='#ffffff', edgecolor='None', boxstyle='round,pad=0.25'))
    ax.text(34, 55, f'Box entries\nfrom center: {len(df_cent)}', fontsize=18, ha='center', va='center', zorder=7,
            bbox=dict(facecolor='#ffffff', edgecolor='None', boxstyle='round,pad=0.25'))
    ax.text(0, 55, f'Box entries\nfrom right: {len(df_rigt)}', fontsize=18, ha='right', va='center', zorder=7,
            bbox=dict(facecolor='#ffffff', edgecolor='None', boxstyle='round,pad=0.25'))
    ax_text(60, 112, s=f'Box entries by open-play Passes: {t}, <Led to shot: {s}>, <Led to goal: {g}>', fontsize=13, color='k', ax=ax,
            highlight_textprops=[{'color':col2}, {'color':'green'}], ha='left', va='center')
    ax.text(60, 109, f'Data from: Opta | Till GW10 in LaLiga 2024-25 | made by: @adnaaan433', fontsize=13, color='k', ha='left', va='center')
    
    IMAGE_URL = f'file:///D:/FData/club_teams_logo/all_league%20-%20Copy/{team_name}.html'
    image = Image.open(urlopen(IMAGE_URL))
    ax_image = pitch.inset_image(113, 68, image, height=10, alpha=1, zorder=1, ax=ax)
    
    return

fig,ax=plt.subplots(figsize=(10,10),facecolor='#f5f5f5')
pass_entry_for(ax,'Real Madrid')
```

```python
df = pd.read_csv("D:\FData\LaLiga_2023-24\GW1_to_GW_38combinded.csv")
```

```python
def final_third_entry_vs_box_entry(team_name):
    final_pass = df[(df['teamName']==team_name) & (df['type']=='Pass') & (df['x']<70) & (df['endX']>=70) & (df['outcomeType']=='Successful') &
                    (~df['qualifiers'].str.contains('CornerTaken|Freekick|ThrowIn'))]
    final_carry = df[(df['teamName']==team_name) & (df['type']=='Carry') & (df['x']<70) & (df['endX']>=70)]
    final_entry = pd.concat([final_pass, final_carry])
    final_third_entry = len(final_entry)

    box_entry = df[(df['teamName']==team_name) & ((df['type']=='Pass')|(df['type']=='Carry')) & (df['outcomeType']=='Successful') & (df['endX']>=88.5) &
                 ~((df['x']>=88.5) & (df['y']>=13.6) & (df['y']<=54.6)) &
                 (df['endY']>=13.6) & (df['endY']<=54.4) & (~df['qualifiers'].str.contains('CornerTaken|Freekick|ThrowIn'))]
    box_entries = len(box_entry)

    box_entry_efficiency = round((box_entries/final_third_entry)*100, 2)
    
    return{"Final_1/3_entries": final_third_entry,
           "Box_entries": box_entries,
           "Final_1/3_entry_efficiency%": box_entry_efficiency
          }

teams = df['teamName'].unique()

# Create a list of dictionaries to store the counts for each player
data = []

for team_name in teams:
    counts = final_third_entry_vs_box_entry(team_name)
    counts['Team'] = team_name
    data.append(counts)

# Convert the list of dictionaries to a DataFrame
counts_df = pd.DataFrame(data)
```

```python
# Sort the DataFrame by 'pr_count' in descending order
counts_df = counts_df.sort_values(by='Final_1/3_entry_efficiency%', ascending=False).reset_index(drop=True)

counts_df
```

```python
def final_third_entry_vs_box_entry_against(team_name):
    final_pass = df[(df['oppositionTeamName']==team_name) & (df['type']=='Pass') & (df['x']<70) & (df['endX']>=70) & (df['outcomeType']=='Successful') &
                    (~df['qualifiers'].str.contains('CornerTaken|Freekick|ThrowIn'))]
    final_carry = df[(df['oppositionTeamName']==team_name) & (df['type']=='Carry') & (df['x']<70) & (df['endX']>=70)]
    final_entry = pd.concat([final_pass, final_carry])
    final_third_entry = len(final_entry)

    box_entry = df[(df['oppositionTeamName']==team_name) & ((df['type']=='Pass')|(df['type']=='Carry')) & (df['outcomeType']=='Successful') & (df['endX']>=88.5) &
                 ~((df['x']>=88.5) & (df['y']>=13.6) & (df['y']<=54.6)) &
                 (df['endY']>=13.6) & (df['endY']<=54.4) & (~df['qualifiers'].str.contains('CornerTaken|Freekick|ThrowIn'))]
    box_entries = len(box_entry)

    box_entry_efficiency = round((box_entries/final_third_entry)*100, 2)
    
    return{"Final_third_entries_conceded": final_third_entry,
           "Box_entries_conceded": box_entries,
           "Final_third_entry_defending_efficiency%": box_entry_efficiency
          }

teams = df['oppositionTeamName'].unique()

# Create a list of dictionaries to store the counts for each player
data = []

for team_name in teams:
    counts = final_third_entry_vs_box_entry_against(team_name)
    counts['Team'] = team_name
    data.append(counts)

# Convert the list of dictionaries to a DataFrame
counts_df = pd.DataFrame(data)
```
