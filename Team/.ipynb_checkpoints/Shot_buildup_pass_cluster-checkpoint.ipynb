```python
def plot_shot_buildup_pass_cluster(ax, team_name):
    df_team = df[(df['teamName'] == team_name) | (df['oppositionTeamName'] == team_name)].reset_index(drop=True)
    df_team = df_team[(~df_team['qualifiers'].str.contains('CornerTaken|FreekickTaken|ThrowIn'))].reset_index(drop=True)

    pitch = Pitch(pitch_type='uefa', pitch_color='#f5f5f5', line_color='k', corner_arcs=True)
    pitch.draw(ax=ax)

    # Collect passes leading to shots
    pass_data = []
    for i in range(len(df_team)):
        if (('Shot' in df_team.loc[i, 'type']) and (df_team.loc[i, 'teamName']==team_name)):
            if i > 0 and 'Pass' in df_team.loc[i - 1, 'type']:
                pass_data.append((df_team.loc[i - 1, 'x'], df_team.loc[i - 1, 'y'], df_team.loc[i - 1, 'endX'], df_team.loc[i - 1, 'endY'], '#ff4b44'))
            if i > 1 and 'Pass' in df_team.loc[i - 2, 'type']:
                pass_data.append((df_team.loc[i - 2, 'x'], df_team.loc[i - 2, 'y'], df_team.loc[i - 2, 'endX'], df_team.loc[i - 2, 'endY'], 'violet'))
            if i > 2 and 'Pass' in df_team.loc[i - 3, 'type']:
                pass_data.append((df_team.loc[i - 3, 'x'], df_team.loc[i - 3, 'y'], df_team.loc[i - 3, 'endX'], df_team.loc[i - 3, 'endY'], 'orange'))

    # Convert to DataFrame
    pass_df = pd.DataFrame(pass_data, columns=['x', 'y', 'endX', 'endY', 'color'])

    # Drop rows with NaN values
    pass_df = pass_df.dropna()

    # Apply KMeans clustering
    for color in ['#ff4b44', 'violet', 'orange']:
        color_passes = pass_df[pass_df['color'] == color]
        if not color_passes.empty:
            # Use a reasonable number of clusters
            n_clusters = min(4, len(color_passes))  # Ensure n_clusters is at most the number of data points
            kmeans = KMeans(n_clusters=n_clusters, random_state=42)
            color_passes['cluster'] = kmeans.fit_predict(color_passes[['x', 'y', 'endX', 'endY']])

            # Find the most populated cluster
            top_cluster = color_passes['cluster'].value_counts().idxmax()
            top_cluster_passes = color_passes[color_passes['cluster'] == top_cluster]

            # Plot the passes in the top cluster
            for _, row in top_cluster_passes.iterrows():
                pitch.lines(row['x'], row['y'], row['endX'], row['endY'], lw=2, transparent=True, comet=True, color=row['color'], 
                            alpha=0.85 if color == '#ff4b44' else (0.85 if color == 'violet' else 0.85), 
                            zorder=4 if color == '#ff4b44' else (3 if color == 'violet' else 2), ax=ax)
                ax.scatter(row['endX'], row['endY'], c=color, zorder=5, edgecolor='k', linewidth=0.75, s=20)
    ax.set_title(f'{team_name}', fontsize=20, fontweight='bold')

fig, ax = plt.subplots(figsize=(10, 10), facecolor='#f5f5f5')
plot_shot_buildup_pass_cluster(ax, 'Girona')
plt.show()
```
