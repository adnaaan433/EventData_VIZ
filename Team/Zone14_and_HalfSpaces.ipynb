```python
def zone_14(ax, team_name):
    team_df = df[(df['teamName']==team_name) & (df['type']=='Pass') & (df['outcomeType']=='Successful') & 
                 (~df['qualifiers'].str.contains('CornerTaken|Freekick'))]
    z14_i = team_df[(team_df['endX']>=70) & (team_df['endX']<=87.5) & (team_df['endY']>=68/3) & (team_df['endY']<=136/3)]
    z14_f = team_df[(team_df['x']>=70) & (team_df['x']<=87.5) & (team_df['y']>=68/3) & (team_df['y']<=136/3)]

    pitch = VerticalPitch(pitch_type='uefa', pitch_color='w', line_color='k', linewidth=2, corner_arcs=True)
    pitch.draw(ax=ax)

    pitch.lines(z14_i.x, z14_i.y, z14_i.endX, z14_i.endY, color=red, comet=True, lw=1, zorder=2, ax=ax)
    pitch.lines(z14_f.x, z14_f.y, z14_f.endX, z14_f.endY, color=blue, comet=True, lw=1, zorder=2, ax=ax)
    
    return

fig,ax=plt.subplots(figsize=(10,10))
zone_14(ax, 'Barcelona')
```
