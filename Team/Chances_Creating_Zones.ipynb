```python
def ccz(ax, team_name):
    cc = df[(df['teamName']==team_name) & (df['qualifiers'].str.contains('KeyPass')) & (~df['qualifiers'].str.contains('Corner|Freekick'))]
    bcc = cc[cc['qualifiers'].str.contains('BigChance')]
    ass = cc[cc['qualifiers'].str.contains('GoalAssist')]

    pitch = VerticalPitch(pitch_type='uefa', corner_arcs=True, line_zorder=2, pitch_color=bg_color, linewidth=2)
    pitch.draw(ax=ax)

    pcmap = LinearSegmentedColormap.from_list("Pearl Earring - 10 colors",  [bg_color, col2], N=20)

    bin_statistic = pitch.bin_statistic(cc.x, cc.y, bins=(6, 5), normalize=False)
    pitch.heatmap(bin_statistic, ax=ax, cmap=pcmap, edgecolors=bg_color)
    pitch.scatter(cc.x, cc.y, s=70, color='None', alpha=0.75, edgecolor=col1, ax=ax)
    pitch.scatter(ass.x, ass.y, s=70, color='green', alpha=0.75, edgecolor=col1, ax=ax)
    labels = pitch.label_heatmap(bin_statistic, color=line_color, fontsize=25, ax=ax, ha='center', va='center', str_format='{:.0f}')

    ax.text(57, 121, f'{team_name}', fontsize=30, fontweight='bold', ha='left', va='center')
    ax.text(57, 116, f'Open-Play Chance Creating Zone', fontsize=13, ha='left', va='center')
    ax_text(57, 112, f'Open-Play  <Chances: {len(cc)}> | BigChances: {len(bcc)} | <Assists: {len(ass)}>', fontsize=13, ha='left', va='center', ax=ax,
            highlight_textprops=[{'color':col1}, {'color':'green'}])
    ax.text(57, 108, f'Data from: Opta | Till GW10 in LaLiga 2024-25 | made by: @adnaaan433', fontsize=12, color='k', ha='left', va='center')

    IMAGE_URL = f'file:///D:/FData/club_teams_logo/all_league%20-%20Copy/{team_name}.html'
    image = Image.open(urlopen(IMAGE_URL))
    ax_image = pitch.inset_image(115, 66, image, height=13, ax=ax)

    name_counts = cc['shortName'].value_counts()
    name_counts_df = name_counts.reset_index()
    name_counts_df.columns = ['name', 'cc_count']
    name_counts_df = name_counts_df.sort_values(by='cc_count', ascending=False)  
    name_counts_df = name_counts_df.reset_index()
    ax_text(-2, 85, f'''<Top Open-Play>
<Chance Creators:>

{name_counts_df.name[0]}:{name_counts_df.cc_count[0]}
{name_counts_df.name[1]}:{name_counts_df.cc_count[1]}
{name_counts_df.name[2]}:{name_counts_df.cc_count[2]}
''', color='k', fontsize=15, va='center', ha='left', highlight_textprops=[{'fontweight':'bold'},{'fontweight':'bold'}], ax=ax)

    name_counts = bcc['shortName'].value_counts()
    name_counts_df = name_counts.reset_index()
    name_counts_df.columns = ['name', 'cc_count']
    name_counts_df = name_counts_df.sort_values(by='cc_count', ascending=False)  
    name_counts_df = name_counts_df.reset_index()

    ax_text(-2, 52.5, f'''<Top Open-Play>
<Big Chance Creators:>

{name_counts_df.name[0]}:{name_counts_df.cc_count[0]}
{name_counts_df.name[1]}:{name_counts_df.cc_count[1]}
{name_counts_df.name[2]}:{name_counts_df.cc_count[2]}
''', color='k', fontsize=15, va='center', ha='left', highlight_textprops=[{'fontweight':'bold'},{'fontweight':'bold'}], ax=ax)

    name_counts = ass['shortName'].value_counts()
    name_counts_df = name_counts.reset_index()
    name_counts_df.columns = ['name', 'cc_count']
    name_counts_df = name_counts_df.sort_values(by='cc_count', ascending=False)  
    name_counts_df = name_counts_df.reset_index()

    ax_text(-2, 20, f'''<Top Open-Play>
<Assisters:>

{name_counts_df.name[0]}:{name_counts_df.cc_count[0]}
{name_counts_df.name[1]}:{name_counts_df.cc_count[1]}
{name_counts_df.name[2]}:{name_counts_df.cc_count[2]}
''', color='k', fontsize=15, va='center', ha='left', highlight_textprops=[{'fontweight':'bold'},{'fontweight':'bold'}], ax=ax)

    return

fig,ax=plt.subplots(figsize=(10,10), facecolor=bg_color)
ccz(ax, 'Barcelona')
```
