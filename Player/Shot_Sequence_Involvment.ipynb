```python
# Function to create the bar chart
def sh_sq_bar(ax, shot_seq_counts):
    # Get top 10 players by total shot sequence involvement
    top10 = shot_seq_counts.head(10).iloc[::-1]

    # Plot horizontal bar chart
    ax.barh(top10['shortName'], top10['Shots'], height=0.75, label='Shot', color=col1)
    ax.barh(top10['shortName'], top10['Shot_Assist'], height=0.75, label='Shot Assist', color=green, left=top10['Shots'])
    ax.barh(top10['shortName'], top10['Buildup_to_shot'], height=0.75, label='Buildup to Shot', color=col2, left=top10[['Shots', 'Shot_Assist']].sum(axis=1))

    # Add counts in the middle of the bars (if count > 0)
    for i, player in enumerate(top10['shortName']):
        for j, count in enumerate(top10[['Shots', 'Shot_Assist', 'Buildup_to_shot']].iloc[i]):
            if count > 0:
                x_position = sum(top10.iloc[i, 1:1+j]) + count / 2
                ax.text(x_position, i, str(count), ha='center', va='center', color=bg_color, fontsize=17, fontweight='bold')
        # Add total count at the end of the bar
        ax.text(top10['total'].iloc[i] + 1, i, str(top10['total'].iloc[i]), ha='left', va='center', color=line_color, fontsize=17, fontweight='bold')
        # Plotting the logos
        for i, row in top10.iterrows():
            img = Image.open(f"D:\\FData\\club_teams_logo\\all_league\\{row['team_names']}.png")
            imagebox = OffsetImage(img, zoom=0.1)  # Adjust zoom as needed
            ab = AnnotationBbox(imagebox, (0, 9-i), frameon=False, xybox=(-40, 0), xycoords='data', boxcoords="offset points")
            ax.add_artist(ab)

    max_x = top10['total'].max()
    x_coord = [10 * i for i in range(1, int(max_x / 10))]
    for x in x_coord:
        ax.axvline(x=x, color='gray', linestyle='--', zorder=2, alpha=0.5)

    ax.set_facecolor(bg_color)
    ax.tick_params(axis='x', colors=line_color, labelsize=15)
    ax.tick_params(axis='y', colors=line_color, labelsize=18, pad=75)
    ax.xaxis.label.set_color(line_color)
    ax.yaxis.label.set_color(line_color)
    for spine in ax.spines.values():
        spine.set_edgecolor(bg_color)

    # ax.set_title(title, color=line_color, fontsize=15, fontweight='bold')
    ax.legend(fontsize=18, loc='lower right')
```

# LaLiga

```python
# Get unique players
unique_players = df['shortName'].unique()

# Shot Sequence Involvement
df_no_carry = df[df['type']!='Carry'].reset_index(drop=True)
# Initialize an empty dictionary to store home players different type of shot sequence counts
shot_seq_counts = {'shortName': unique_players, 'Shots': [], 'Shot_Assist': [], 'Buildup_to_shot': [], 'team_names': []}
# Putting counts in those lists
for name in unique_players:
    dfp = df_no_carry[df_no_carry['shortName'] == name]
    shot_seq_counts['Shots'].append(len(dfp[dfp['type'].isin(['MissedShots','SavedShot','ShotOnPost','Goal'])]))
    shot_seq_counts['Shot_Assist'].append(len(dfp[(dfp['qualifiers'].str.contains('KeyPass'))]))
    shot_seq_counts['Buildup_to_shot'].append(len(df_no_carry[(df_no_carry['type'] == 'Pass') & (df_no_carry['outcomeType']=='Successful') & 
                                                              (df_no_carry['shortName'] == name) &
                                                              (df_no_carry['qualifiers'].shift(-1).str.contains('KeyPass'))]))
    
    shot_seq_counts['team_names'].append(dfp['teamName'].max())
# converting that list into a dataframe
sh_sq_df = pd.DataFrame(shot_seq_counts)
sh_sq_df['total'] = sh_sq_df['Shots']+sh_sq_df['Shot_Assist']+sh_sq_df['Buildup_to_shot']
sh_sq_df = sh_sq_df.sort_values(by='total', ascending=False)
sh_sq_df.reset_index(drop=True, inplace=True)
sh_sq_df = sh_sq_df.head(10)

# Plotting
fig, ax = plt.subplots(figsize=(16, 16), facecolor='None')

sh_sq_bar(ax, sh_sq_df)

# # Set the overall title for the figure
fig.text(0, 0.95, "LaLiga Players Shot Sequences Involvments", color='w', fontsize=40, fontweight='bold', ha='left')
fig.text(0, 0.92, "Data from: Opta | Till GW11, LaLiga 2024-25 | made by: @adnaaan433", color='w', fontsize=20, ha='left')

# # plt.tight_layout(rect=[0, 0, 1, 0.96])
# plt.show()

fig.savefig(f"D:\FData\LaLiga_2024_25\Analysis\Shot_Seq.png", bbox_inches='tight')
```
