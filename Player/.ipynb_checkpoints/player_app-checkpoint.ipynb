```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
import matplotlib.patches as patches
import matplotlib as mpl
from mplsoccer import Pitch, VerticalPitch, FontManager, Sbopen, add_image, Radar
from matplotlib.font_manager import FontProperties
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.patheffects as path_effects
import matplotlib.gridspec as gridspec
from highlight_text import ax_text, fig_text
from PIL import Image
from mplsoccer import add_image
from urllib.request import urlopen
import os
from unidecode import unidecode


# Print the modified DataFrame
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

green = '#69f900'
red = '#ff4b44'
blue = '#00a0de'
violet = '#a369ff'
bg_color= '#f5f5f5'
line_color= '#000000'
hcol = '#ff4b44'
acol = '#00a0de'
```

```python
df = pd.read_csv(r"D:\FData\2023_24_season_data\EventData(EPL_and_LaLiga).csv")
df_shots = pd.read_csv(r"D:\FData\2023_24_season_data\ShotsData(EPL_and_LaLiga).csv")
```

```python
def All_Pass(ax, pname):
    dfp = df[(df['name']==pname) & (df['type']=='Pass')]
    dfs = dfp[dfp['outcomeType']=='Successful']
    dfu = dfp[dfp['outcomeType']=='Unsuccessful']
    
    pitch = VerticalPitch(pitch_type='uefa', corner_arcs=True, pitch_color=bg_color, line_color=line_color, linewidth=2)
    pitch.draw(ax=ax)
    ax.set_xlim(-0.5, 68.5)
    ax.set_ylim(-0.5, 105.5)
    ax.invert_xaxis()

    pitch.lines(dfs.x, dfs.y, dfs.endX, dfs.endY, color=acol, lw=1, alpha=0.75,comet=True, zorder=4, ax=ax)
    pitch.lines(dfu.x, dfu.y, dfu.endX, dfu.endY, color=hcol, lw=1, alpha=0.5, comet=True, zorder=3, ax=ax)

    pitch.scatter(dfs.endX, dfs.endY, s=20, color=bg_color,  edgecolor= acol,  alpha=0.75, zorder=4, ax=ax)
    pitch.scatter(dfu.endX, dfu.endY, s=25, color=bg_color,  edgecolor=hcol, alpha=0.5, zorder=3, ax=ax)

    team_name = dfp['teamName'].max()
    position = dfp['position'].max()
    IMAGE_URL = f'file:///D:/FData/club_teams_logo/all_league%20-%20Copy/{team_name}.html'
    image = Image.open(urlopen(IMAGE_URL))
    ax_image = pitch.inset_image(112, 61, image, height=8, ax=ax)
    ax.text(53, 113, f'{pname}', color='k', fontsize=20, fontweight='bold', ha='left', va='center')
    ax.text(53, 109, f'in {team_name}, 23-24 season', color='k', fontsize=15, ha='left', va='center')
    ax.text(34, -2, f'Total Pass: {len(dfp)}', fontsize=15, ha='center', va='center')
    ax_text(34, -5, f'<Successful: {len(dfs)}>    |    <Unsuccessul: {len(dfu)}>', fontsize=15, 
            highlight_textprops=[{'color': acol}, {'color': hcol}], ha='center', va='center', ax=ax)


def OpenPlay_Pass(ax, pname):
    dfp = df[(df['name']==pname) & (df['type']=='Pass') & (df['qualifiers'].str.contains('CornerTaken|FreekickTaken|GoalKick'))]
    dfs = dfp[dfp['outcomeType']=='Successful']
    dfu = dfp[dfp['outcomeType']=='Unsuccessful']
    
    pitch = VerticalPitch(pitch_type='uefa', corner_arcs=True, pitch_color=bg_color, line_color=line_color, linewidth=2)
    pitch.draw(ax=ax)
    ax.set_xlim(-0.5, 68.5)
    ax.set_ylim(-0.5, 105.5)
    ax.invert_xaxis()

    pitch.lines(dfs.x, dfs.y, dfs.endX, dfs.endY, color=acol, lw=1, alpha=0.75,comet=True, zorder=4, ax=ax)
    pitch.lines(dfu.x, dfu.y, dfu.endX, dfu.endY, color=hcol, lw=1, alpha=0.5, comet=True, zorder=3, ax=ax)

    pitch.scatter(dfs.endX, dfs.endY, s=20, color=bg_color,  edgecolor= acol,  alpha=0.75, zorder=4, ax=ax)
    pitch.scatter(dfu.endX, dfu.endY, s=25, color=bg_color,  edgecolor=hcol, alpha=0.5, zorder=3, ax=ax)

    team_name = dfp['teamName'].max()
    position = dfp['position'].max()
    IMAGE_URL = f'file:///D:/FData/club_teams_logo/all_league%20-%20Copy/{team_name}.html'
    image = Image.open(urlopen(IMAGE_URL))
    ax_image = pitch.inset_image(112, 61, image, height=8, ax=ax)
    ax.text(53, 113, f'{pname}', color='k', fontsize=20, fontweight='bold', ha='left', va='center')
    ax.text(53, 109, f'in {team_name}, 23-24 season', color='k', fontsize=15, ha='left', va='center')
    ax_text(34, -5, f'<Successful: {len(dfs)}>    |    <Unsuccessul: {len(dfu)}>', fontsize=15, 
            highlight_textprops=[{'color': acol}, {'color': hcol}], ha='center', va='center', ax=ax)
    

fig,ax=plt.subplots(figsize=(16,12))
All_Pass(ax, 'Bruno Fernandes')
```
